### Polynomial regression

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
import pandas as pd
```

```python
np.random.seed(42)
#This method is called when `RandomState` is initialized.
n_samples = 100

X = np.linspace(0,10,100)
c = np.random.randn(100)

y = X**3+c*100+100

plt.figure(figsize=(8,6))
plt.scatter(X,y)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
lr.fit(X.reshape(-1,1),y)

model_pred = lr.predict(X.reshape(-1,1))

plt.figure(figsize=(8,6))
plt.scatter(X,y)
plt.plot(X,model_pred)
print r2_score(y,model_pred)
```

#### here linear regression is not able to well fitted all data points

### Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree=2)
x_poly = poly_reg.fit_transform(X.reshape(-1,1))
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
y_pred = lin_reg2.predict(x_poly)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X,y)
plt.plot(X,y_pred)
print r2_score(y,y_pred)
```
